```python
import cv2
from PIL import Image, ImageDraw, ImageFont
from matplotlib import pyplot as plt
```

```python
# load image

# OpenCV
# 
# imread(filename[,flag])
#  filename :  
#  flag : 
# 
#    1, cv2.IMREAD_COLOR ()
#    0, cv2.IMREAD_GRAYSCALE
#   -1, cv2.IMREAD_UNCHANGED
img_cv = cv2.imread('IMGP4209.JPG')

# Pillow
img_pl = Image.open('IMGP4209.JPG') #.convert('L')

h, w = img_cv.shape[:2]

w_resize = int(w * 0.8)
h_resize = int(h * 0.8)

# 
img_cv = cv2.resize(img_cv, (w_resize, h_resize))
img_pl = img_pl.resize((w_resize, h_resize))

plt.figure(figsize=(15, 15))
plt.subplot(1, 2, 1), plt.imshow(cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB)) # , , 
plt.subplot(1, 2, 2), plt.imshow(img_pl), plt.show()

# 
# quality : 19575
#   100JPEG
cv2.imwrite('img_cv.jpg', img_cv)
img_pl.save('img_pl.jpg', quality=95) 
```
