```python
import cv2
from PIL import Image, ImageDraw, ImageFont
from matplotlib import pyplot as plt
```

```python
# load image

# OpenCV
# 
# imread(filename[,flag])
#  filename : 読み込み対象ファイル 
#  flag : 画像の読み込み方法を指定するフラグ
# 
#    1, cv2.IMREAD_COLOR (デフォルト。カラーモードで読み込む。アルファチャンネルは読み込まない)
#    0, cv2.IMREAD_GRAYSCALE（グレースケールで読み込む）
#   -1, cv2.IMREAD_UNCHANGED（画像をそのまま読み込む。アルファチャンネルも読み込む）
img_cv = cv2.imread('IMGP4209.JPG')

# Pillow
img_pl = Image.open('IMGP4209.JPG') #.convert('L')

h, w = img_cv.shape[:2]

w_resize = int(w * 0.8)
h_resize = int(h * 0.8)

# リサイズ
img_cv = cv2.resize(img_cv, (w_resize, h_resize))
img_pl = img_pl.resize((w_resize, h_resize))

plt.figure(figsize=(15, 15))
plt.subplot(1, 2, 1), plt.imshow(cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB)) # 行数, 列数, 何番目の
plt.subplot(1, 2, 2), plt.imshow(img_pl), plt.show()

# 保存
# quality : 画質は、1（最悪）から95（最高）までのスケールで指定する。デフォルトは75
#   　　　　100を指定しても、JPEG圧縮アルゴリズムの一部を無効にし、画質の向上がほとんどない大きなファイルを生成する
cv2.imwrite('img_cv.jpg', img_cv)
img_pl.save('img_pl.jpg', quality=95) 
```
